# ğŸŒˆ Mood Tracker App (Tkinter)
Click a button to record your mood.

```
import tkinter as tk
from datetime import datetime

# Function to log mood
def log_mood(mood):
    with open("melisa_moods_gui.txt", "a") as f:
        f.write(f"{datetime.now()}: {mood}\n")
    label.config(text=f"Feeling {mood} logged! ğŸŒŸ")

# GUI setup
root = tk.Tk()
root.title("Melisa's Mood Tracker")

label = tk.Label(root, text="How are you feeling today?", font=('Arial', 14))
label.pack(pady=10)

moods = ['Happy ğŸ˜Š', 'Excited ğŸ‰', 'Tired ğŸ˜´', 'Chill ğŸ§˜â€â™€ï¸']
for m in moods:
    mood = m.split()[0]
    tk.Button(root, text=m, command=lambda m=mood: log_mood(m), font=('Arial', 12)).pack(pady=5)

root.mainloop()
```
