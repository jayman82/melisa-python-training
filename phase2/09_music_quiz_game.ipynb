Use lists and scoring to create a fun quiz!

```
questions = [
    ("Who sings 'Shake It Off'?", 'taylor swift'),
    ("Finish the lyric: 'Let it...'", 'go')
]

score = 0
for q, a in questions:
    guess = input(q + ' ').lower()
    if guess == a:
        print('Correct! ðŸŽ¶')
        score += 1
    else:
        print('Oops!')
print(f'Your final score: {score}/{len(questions)}')
```
