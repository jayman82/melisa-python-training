# ⭐ Catch the Star Game (PyGame)
Use arrow keys to move and catch stars!

```
# This game uses PyGame, which works best outside notebooks. Save this as a `.py` file if needed.
import pygame
import random
import sys

# Initialize PyGame
pygame.init()

# Screen setup
WIDTH, HEIGHT = 500, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Melisa's Catch the Star ⭐")

# Colors
WHITE = (255, 255, 255)
BLUE = (50, 150, 255)
YELLOW = (255, 255, 0)

# Player setup
player_size = 50
player = pygame.Rect(WIDTH//2, HEIGHT - player_size - 10, player_size, player_size)

# Star setup
star_size = 30
star = pygame.Rect(random.randint(0, WIDTH - star_size), 0, star_size, star_size)
star_speed = 5

# Score
score = 0
font = pygame.font.SysFont("Arial", 24)

# Game loop
clock = pygame.time.Clock()
running = True

while running:
    screen.fill(WHITE)
    pygame.draw.rect(screen, BLUE, player)
    pygame.draw.rect(screen, YELLOW, star)

    text = font.render(f"Score: {score}", True, (0, 0, 0))
    screen.blit(text, (10, 10))

    star.y += star_speed
    if star.y > HEIGHT:
        star.y = 0
        star.x = random.randint(0, WIDTH - star_size)

    if player.colliderect(star):
        score += 1
        star.y = 0
        star.x = random.randint(0, WIDTH - star_size)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Player movement
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player.x > 0:
        player.x -= 7
    if keys[pygame.K_RIGHT] and player.x < WIDTH - player_size:
        player.x += 7

    pygame.display.flip()
    clock.tick(30)

pygame.quit()
sys.exit()
```
