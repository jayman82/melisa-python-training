# ðŸŽ‰ Cheer Clicker (Tkinter)
Click the button to get a cheer!

```
import tkinter as tk

# Function to update the label when the button is clicked
def cheer():
    count.set(count.get() + 1)
    cheer_label.config(text=f"Go Melisa! ðŸŽ‰ Clicks: {count.get()}")

# Main window
root = tk.Tk()
root.title("Melisa's Cheer Clicker")

count = tk.IntVar(value=0)

cheer_label = tk.Label(root, text="Click to cheer!", font=('Arial', 16))
cheer_label.pack(pady=10)

cheer_button = tk.Button(root, text="Cheer!", command=cheer, font=('Arial', 14), bg='pink')
cheer_button.pack(pady=10)

root.mainloop()
```
