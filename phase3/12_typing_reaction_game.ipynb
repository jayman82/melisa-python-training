# ⏱️ Typing Reaction Game
Test how fast you can type a random phrase!

```
import time, random

phrases = ['Python is fun', 'Go Melisa!', 'Cheer your heart out']
target = random.choice(phrases)
print("Type this: ", target)
start = time.time()
typed = input(">> ")
end = time.time()

if typed == target:
    print(f'Great job! Time: {end - start:.2f} seconds')
else:
    print("Oops! Try again.")
```
